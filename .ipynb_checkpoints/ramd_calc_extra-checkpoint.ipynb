<p style="color: brown; font-size: 24px;"><strong><em>2) Procedure of approximation and its interpretation</em></strong><br><span style="font-size: 18px; font-style: italic;">Data loading and initial processing</span></p>

<p style="color: darkblue; font-size: 24px;"><strong><em> force 250 : </em></strong></p>

```python
force = forces_data['force250']
force_value = 250
tau_frac = 15500

# Filtering
t, N = filter_data(force)
```

<span style="color:darkblue;font-size: 18px;"><em>a) Finding classical exponential relaxation:</em></span>

```python
N0 = N[0]
n = np.log(N / N0)
tau_var = []  # Для значений времени диссоциации
pend_var = []  # Для коэффициентов линейного приближения
lambda_var = []  # Для значений скорости диссоциации
```

```python
# Перебираем значения N_cut и рассчитываем tau, lambda и pend для каждого
for N_cut in range(len(N)):
    # Фильтруем данные для текущего N_cut и учитываем tau_frac
    t_fit = t[(N > N_cut) & (t > tau_frac)]  # Оставляем только те точки, где N > N_cut и t > tau_frac
    n_fit = n[(N > N_cut) & (t > tau_frac)]  # Оставляем только соответствующие значения логарифма вероятности выживания

    # Проверка на непустые t_fit и n_fit для работы polyfit
    if len(t_fit) > 0 and len(n_fit) > 0:
        # Линейное приближение
        pend = np.polyfit(t_fit, n_fit, 1)
        
        # Рассчитываем tau и lambda
        tau = (1 / abs(pend[0])) / 1e4  # Расчет tau с масштабированием на 1e4
        lambda_ = pend[0] * 1e4         # Расчет lambda с масштабированием на 1e4
        
        # Сохраняем результаты в соответствующие списки
        tau_var.append(tau)
        lambda_var.append(lambda_)
        pend_var.append(pend)
    else:
        # Заполняем None для пропуска значений, где данных недостаточно
        tau_var.append(None)
        lambda_var.append(None)
        pend_var.append(None)

# Фильтрация None значений для построения графика
filtered_tau_var = [val for val in tau_var if val is not None]
filtered_N_cut_range = [i for i in range(len(tau_var)) if tau_var[i] is not None]

# Построение графика зависимости tau от N_cut
plt.figure(figsize=(8, 4))
plt.plot(filtered_N_cut_range, filtered_tau_var, label=r'$\tau$ values', marker='o')
plt.xlabel(r'$N_{cut}$')
plt.ylabel(r'$\tau$, ps')
plt.title("Tau vs N_cut for All Data with tau_frac Filter")
plt.grid(True)
plt.legend()
plt.show()
```
