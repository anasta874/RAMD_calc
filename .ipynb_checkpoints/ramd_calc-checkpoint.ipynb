```python
force = forces_data['force250']
force_value = 250

# Фильтрация
valid_ind = (~np.isnan(force[:, 1])) & (force[:, 1] > 0)  
t = force[valid_ind, 0]
N = force[valid_ind, 1]

# # Вывод данных для проверки
# print("Время (t):", t)
# print("Величины сил (N):", N)
```

<span style="color:darkblue;font-size: 18px;"><em>а) Поиск классической экпоненциальной релаксации:</em></span>

```python
# создаем пустые массивы, которые будут заполняться значениями
tau_all = [] #список всех времен диссоциации
forces_all = [] # список сил
pend_var = [] # для коэффициентов линейной аппроксимации
tau_var = [] # для хранения значений времени диссоциации
lambda_var = [] # значения скорости диссоциации
```

Время диссоциации (tau) — это среднее время, которое требуется для того, чтобы лиганд отсоединился от белка

```python
tau_frac = 15500
N_cut_max = 13
```

```python
N0 = N[0] #берет начальное значение количества связанных лигандов
# n = np.log(N0 / N) #логарифм отношения начального количества лигандов к текущему количеству
n = np.log(N / N0)
```

```python
for N_cut in range(N_cut_max):
    t_fit = []
    n_fit = []
    
    # Фильтрация данных: отбираем значения, которые больше N_cut и tau_frac
    for i in range(len(t)):
        if N[i] > N_cut and t[i] > tau_frac:
            t_fit.append(t[i])
            n_fit.append(n[i])
            
    t_fit = [value for value in t_fit if value > 0] # Фильтрация положительных значений для t_fit
    pend = np.polyfit(t_fit, n_fit, 1)
    tau = (1 / abs(pend[0])) / 1e4  # Делим на 10^4 для масштабирования
    lambda_ = pend[0] * 1e4  # Умножаем на 10^4 для масштабирования

    tau_var.append(round(tau, 15)) 
    lambda_var.append(round(lambda_, 15)) 
    pend_var.append(pend)

# Вывод для проверки
print("tau_var:", tau_var)
print("lambda_var:", lambda_var)
```

```python
plt.plot(range(13), tau_var[:13], label='tau (ps)')
plt.xlabel('N_cut')
plt.ylabel('tau (ps)')
plt.grid(True)
plt.legend()
plt.show()
```
